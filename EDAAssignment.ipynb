```python

# 11. Identify and remove outliers in the km_driven column using the IQR method.

Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter dataset to remove outliers
df_no_outliers = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]

# Number of records removed
outliers_removed = df.shape[0] - df_no_outliers.shape[0]

outliers_removed


```

```python
# 12.  Perform a bivariate analysis to visualize the relationship between year and selling_price


plt.figure(figsize=(10, 5))
sns.boxplot(x=df_no_outliers["year"], y=df_no_outliers["selling_price"])
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price (INR)")
plt.title("Selling Price Distribution Across Manufacturing Years")
plt.xticks(rotation=45)
plt.show()
```

```python
# 13.  What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)

from datetime import datetime

current_year = datetime.now().year

# Calculate bike age
df_no_outliers["bike_age"] = current_year - df_no_outliers["year"]

# Compute average selling price per age
avg_depreciation = df_no_outliers.groupby("bike_age")["selling_price"].mean()

avg_depreciation
```

```python
# 14. Which bike names are priced significantly above the average price for their manufacturing year?

# Compute mean and standard deviation for each year
yearly_avg = df_no_outliers.groupby("year")["selling_price"].mean()
yearly_std = df_no_outliers.groupby("year")["selling_price"].std()

# Define threshold: Mean + 1.5 * Std Dev
thresholds = yearly_avg + 1.5 * yearly_std

# Identify bikes priced above the threshold
high_priced_bikes = df_no_outliers[df_no_outliers.apply(lambda x: x["selling_price"] > thresholds[x["year"]], axis=1)]

# Select relevant columns
high_priced_bikes[["name", "year", "selling_price"]]

```

```python
# 15.Develop a correlation matrix for numeric columns and visualize it using a heatmap.

import numpy as np

numeric_cols = df_no_outliers.select_dtypes(include=[np.number])
correlation_matrix = numeric_cols.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features")
plt.show()

```

```python
# Reload necessary libraries
import pandas as pd

# Load the car sale dataset
car_file_path = "/mnt/data/Car Sale - Car Sale.csv"
car_df = pd.read_csv(car_file_path)

# Display basic info and first few rows
car_df.info(), car_df.head()


# 1. Average selling price of cars for each dealer

dealer_avg_price = car_df.groupby("Dealer_Name")["Price ($)"].mean().sort_values(ascending=False)

# Display top 10 dealers with highest average selling price
dealer_avg_price.head(10)


# 2. Car brand with the highest variation in prices

brand_price_variation = car_df.groupby("Company")["Price ($)"].std().sort_values(ascending=False)

# Display top 10 brands with the highest standard deviation in price
brand_price_variation.head(10)

import matplotlib.pyplot as plt
import seaborn as sns

# 3. Distribution of car prices for each transmission type

plt.figure(figsize=(10, 5))
sns.boxplot(x=car_df["Transmission"], y=car_df["Price ($)"], palette="coolwarm")
plt.xlabel("Transmission Type")
plt.ylabel("Price ($)")
plt.title("Distribution of Car Prices by Transmission Type")
plt.yscale("log")  # Log scale to handle skewed price distribution
plt.show()

# 4. Distribution of car prices across different regions

plt.figure(figsize=(12, 6))
sns.boxplot(x=car_df["Dealer_Region"], y=car_df["Price ($)"], palette="viridis")
plt.xlabel("Dealer Region")
plt.ylabel("Price ($)")
plt.title("Distribution of Car Prices by Region")
plt.xticks(rotation=45)
plt.yscale("log")  # Log scale for better visualization of price differences
plt.show()


# 5. Distribution of cars based on body styles

plt.figure(figsize=(12, 6))
sns.countplot(y=car_df["Body Style"], order=car_df["Body Style"].value_counts().index, palette="magma")
plt.xlabel("Count of Cars")
plt.ylabel("Body Style")
plt.title("Distribution of Cars by Body Style")
plt.show()


# 6. Average selling price of cars by customer gender

gender_avg_price = car_df.groupby("Gender")["Price ($)"].mean()

# 6.1 Plot gender-wise average price
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_avg_price.index, y=gender_avg_price.values, palette="coolwarm")
plt.xlabel("Gender")
plt.ylabel("Average Car Price ($)")
plt.title("Average Selling Price by Customer Gender")
plt.show()

# 6.2 Relationship between Annual Income and Car Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x=car_df["Annual Income"], y=car_df["Price ($)"], alpha=0.5, color="blue")
plt.xlabel("Annual Income ($)")
plt.ylabel("Car Price ($)")
plt.title("Car Price vs. Customer Annual Income")
plt.xscale("log")  # Log scale to manage income variation
plt.yscale("log")  # Log scale for price distribution
plt.show()


# 7. Distribution of car prices and number of cars sold by region

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot of car prices by region
sns.boxplot(x=car_df["Dealer_Region"], y=car_df["Price ($)"], palette="viridis", ax=ax[0])
ax[0].set_xlabel("Region")
ax[0].set_ylabel("Price ($)")
ax[0].set_title("Car Price Distribution by Region")
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)
ax[0].set_yscale("log")

# Count of cars sold by region
region_sales = car_df["Dealer_Region"].value_counts()
sns.barplot(x=region_sales.index, y=region_sales.values, palette="magma", ax=ax[1])
ax[1].set_xlabel("Region")
ax[1].set_ylabel("Number of Cars Sold")
ax[1].set_title("Number of Cars Sold by Region")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()


# 8. Average car price by engine size

engine_avg_price = car_df.groupby("Engine")["Price ($)"].mean().sort_values(ascending=False)

# Plot engine size vs. average price
plt.figure(figsize=(12, 6))
sns.barplot(x=engine_avg_price.index, y=engine_avg_price.values, palette="coolwarm")
plt.xlabel("Engine Size")
plt.ylabel("Average Car Price ($)")
plt.title("Average Car Price by Engine Size")
plt.xticks(rotation=45)
plt.show()


# 9. Car prices based on annual income brackets

# Define income brackets
income_bins = [0, 30000, 60000, 100000, 150000, 200000, car_df["Annual Income"].max()]
income_labels = ["<30K", "30K-60K", "60K-100K", "100K-150K", "150K-200K", "200K+"]
car_df["Income Bracket"] = pd.cut(car_df["Annual Income"], bins=income_bins, labels=income_labels)

# Plot car price distribution by income bracket
plt.figure(figsize=(12, 6))
sns.boxplot(x=car_df["Income Bracket"], y=car_df["Price ($)"], palette="coolwarm")
plt.xlabel("Annual Income Bracket ($)")
plt.ylabel("Car Price ($)")
plt.title("Car Price Distribution by Annual Income Bracket")
plt.yscale("log")  # Log scale for better visualization
plt.show()


# 10. Top 5 car models with the highest number of sales

top_models = car_df["Model"].value_counts().head(5).index

# Filter dataset for only top 5 models
top_models_df = car_df[car_df["Model"].isin(top_models)]

# Boxplot of price distribution for top 5 models
plt.figure(figsize=(12, 6))
sns.boxplot(x=top_models_df["Model"], y=top_models_df["Price ($)"], palette="magma")
plt.xlabel("Car Model")
plt.ylabel("Car Price ($)")
plt.title("Price Distribution for Top 5 Car Models")
plt.yscale("log")
plt.show()


# 11. Car price variation by engine size and color

plt.figure(figsize=(14, 6))
sns.boxplot(x=car_df["Color"], y=car_df["Price ($)"], hue=car_df["Engine"], palette="viridis")
plt.xlabel("Car Color")
plt.ylabel("Car Price ($)")
plt.title("Car Price Variation by Color and Engine Size")
plt.xticks(rotation=45)
plt.yscale("log")  # Log scale for better visualization
plt.legend(title="Engine Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()


# 12. Seasonal trend in car sales

# Convert Date column to datetime format
car_df["Date"] = pd.to_datetime(car_df["Date"])

# Extract month and group by sales count
car_df["Month"] = car_df["Date"].dt.month
monthly_sales = car_df["Month"].value_counts().sort_index()

# Plot monthly sales trend
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o", color="blue")
plt.xlabel("Month")
plt.ylabel("Number of Cars Sold")
plt.title("Seasonal Trend in Car Sales")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid()
plt.show()


# 13. Car price distribution by body style and transmission type

plt.figure(figsize=(14, 6))
sns.boxplot(x=car_df["Body Style"], y=car_df["Price ($)"], hue=car_df["Transmission"], palette="coolwarm")
plt.xlabel("Body Style")
plt.ylabel("Car Price ($)")
plt.title("Car Price Distribution by Body Style and Transmission Type")
plt.xticks(rotation=45)
plt.yscale("log")  # Log scale for better visualization
plt.legend(title="Transmission", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()


# 14. Correlation matrix for car price, engine size, and annual income

# Select relevant numeric features and encode engine types as numeric categories
car_df["Engine_Category"] = car_df["Engine"].astype("category").cat.codes

# Compute correlation matrix
corr_matrix = car_df[["Price ($)", "Engine_Category", "Annual Income"]].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Car Price, Engine Size, and Annual Income")
plt.show()



# 15. Average car price by car model and engine type

avg_price_model_engine = car_df.groupby(["Model", "Engine"])["Price ($)"].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.scatterplot(data=avg_price_model_engine, x="Model", y="Price ($)", hue="Engine", palette="magma", s=100)
plt.xlabel("Car Model")
plt.ylabel("Average Car Price ($)")
plt.title("Average Car Price by Model and Engine Type")
plt.xticks(rotation=90)
plt.yscale("log")
plt.legend(title="Engine Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()


```

 1.What is the average rating for each product category?

 2.What are the top rating_count products by category?

 3.What is the distribution of discounted prices vs. actual prices?

 5.What are the most popular product names?

 6.What are the most popular product keywords?

 7.What are the most popular product reviews?

 10.Identify any potential areas for improvement or optimization based on the data analysis.

```python
import pandas as pd

# Load the dataset
file_path = "/mnt/data/amazon - amazon.csv"
df = pd.read_csv(file_path)

# Display basic information and first few rows
df.info(), df.head()


# Data Cleaning

# Convert price columns to numeric values (removing currency symbols and commas)
df["discounted_price"] = df["discounted_price"].replace('[₹,]', '', regex=True).astype(float)
df["actual_price"] = df["actual_price"].replace('[₹,]', '', regex=True).astype(float)

# Convert discount percentage to numeric values (removing % sign)
df["discount_percentage"] = df["discount_percentage"].replace('%', '', regex=True).astype(float)

# Convert rating to numeric values
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')

# Convert rating_count to numeric values (removing commas)
df["rating_count"] = df["rating_count"].replace(',', '', regex=True).astype(float)

# Extract the main category (before the first "|") for better analysis
df["main_category"] = df["category"].apply(lambda x: x.split('|')[0] if isinstance(x, str) else x)

# Display cleaned dataset info
df.info(), df.head()




# 1. Average rating for each product category
avg_rating_per_category = df.groupby("main_category")["rating"].mean().sort_values(ascending=False)

# 2. Top rating_count products by category
top_rated_products_by_category = df.loc[df.groupby("main_category")["rating_count"].idxmax(),
                                        ["main_category", "product_name", "rating_count"]]

# 3. Distribution of discounted prices vs. actual prices (summary statistics)
price_distribution = df[["discounted_price", "actual_price"]].describe()

# 4. Average discount percentage across categories
avg_discount_per_category = df.groupby("main_category")["discount_percentage"].mean().sort_values(ascending=False)

# 5. Most popular product names (by count of reviews)
popular_products = df["product_name"].value_counts().head(10)

# 6. Most popular product keywords (based on word frequency in product names)
from collections import Counter

word_counts = Counter(" ".join(df["product_name"]).lower().split())
popular_keywords = pd.DataFrame(word_counts.most_common(20), columns=["Keyword", "Count"])

# 7. Most popular product reviews (based on length and rating count)
df["review_length"] = df["review_content"].apply(lambda x: len(str(x)))
popular_reviews = df.sort_values(["rating_count", "review_length"], ascending=[False, False]) \
                    [["product_name", "review_content", "rating_count"]].head(10)

# 8. Correlation between discounted_price and rating
correlation_price_rating = df[["discounted_price", "rating"]].corr().iloc[0, 1]

# 9. Top 5 categories based on highest ratings
top_categories_by_rating = avg_rating_per_category.head(5)

# Compile all results into a dictionary
analysis_results = {
    "Average Rating per Category": avg_rating_per_category,
    "Top Rated Products by Category": top_rated_products_by_category,
    "Price Distribution Summary": price_distribution,
    "Average Discount Percentage per Category": avg_discount_per_category,
    "Most Popular Products": popular_products,
    "Most Popular Product Keywords": popular_keywords,
    "Most Popular Product Reviews": popular_reviews,
    "Correlation between Discounted Price and Rating": correlation_price_rating,
    "Top 5 Categories by Highest Ratings": top_categories_by_rating
}

analysis_results




"""
Potential Areas for Improvement
Car & Motorbike category has the lowest rating (3.80). Identifying product-specific issues could help.

Toys & Games have no recorded discounts – implementing offers might boost sales.

Weak correlation between discounted price and rating – discounts may not be driving better user satisfaction.

Optimize high-rated, high-review products like AmazonBasics HDMI cables and SanDisk USBs by promoting them more.
"""
```

```python
# Load the Spotify dataset
spotify_df = pd.read_csv("/mnt/data/spotify.csv")

# Check for null values
null_values = spotify_df.isnull().sum()

# Check for duplicate rows
duplicate_rows = spotify_df.duplicated().sum()

# Drop duplicates if present
if duplicate_rows > 0:
    spotify_df = spotify_df.drop_duplicates()

# Fill or drop missing values if present
if null_values.any():
    spotify_df = spotify_df.dropna()

# Display the cleaned dataset info
spotify_df.info(), null_values, duplicate_rows



import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of popularity among tracks (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(spotify_df["Popularity"], bins=20, kde=True, color="blue")
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Count")
plt.show()

# 2. Scatter plot of Popularity vs Duration
plt.figure(figsize=(8, 5))
sns.scatterplot(data=spotify_df, x="Duration (ms)", y="Popularity", alpha=0.6, color="green")
plt.title("Relationship Between Track Popularity and Duration")
plt.xlabel("Duration (ms)")
plt.ylabel("Popularity")
plt.show()

# 3. Countplot of tracks per artist
plt.figure(figsize=(10, 5))
top_artists = spotify_df["Artist"].value_counts().head(10)  # Top 10 artists with most tracks
sns.barplot(x=top_artists.index, y=top_artists.values, palette="mako")
plt.xticks(rotation=45)
plt.title("Top Artists with Most Tracks in Dataset")
plt.xlabel("Artist")
plt.ylabel("Number of Tracks")
plt.show()


# 4. Top 5 least popular tracks
least_popular_tracks = spotify_df.nsmallest(5, "Popularity")[["Artist", "Track Name", "Popularity"]]
least_popular_tracks


# 5. Top 5 most popular artists based on average popularity
top_5_popular_artists = spotify_df.groupby("Artist")["Popularity"].mean().nlargest(5)

# Display the top artist with highest average popularity
top_5_popular_artists

# 6. Most popular track for each of the top 5 artists
most_popular_tracks = spotify_df.loc[spotify_df.groupby("Artist")["Popularity"].idxmax(), ["Artist", "Track Name", "Popularity"]]
most_popular_tracks = most_popular_tracks[most_popular_tracks["Artist"].isin(top_5_popular_artists.index)]

most_popular_tracks


# 7. Pair plot to visualize relationships between numerical variables
sns.pairplot(spotify_df[["Popularity", "Duration (ms)"]], diag_kind="kde", corner=True)
plt.show()


# Re-attempt pair plot ensuring numeric columns are properly formatted
numeric_cols = ["Popularity", "Duration (ms)"]
spotify_numeric_df = spotify_df[numeric_cols]

# Generate pair plot
sns.pairplot(spotify_numeric_df, diag_kind="kde", corner=True)
plt.show()


# Check data types of columns
spotify_df.dtypes


# Ensure numeric columns are properly formatted before generating the pair plot
spotify_df["Popularity"] = pd.to_numeric(spotify_df["Popularity"], errors="coerce")
spotify_df["Duration (ms)"] = pd.to_numeric(spotify_df["Duration (ms)"], errors="coerce")

# Generate pair plot for numerical variables
sns.pairplot(spotify_df[["Popularity", "Duration (ms)"]], diag_kind="kde", corner=True)
plt.show()


# Check for non-numeric values in the numerical columns
invalid_popularity = spotify_df[~spotify_df["Popularity"].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
invalid_duration = spotify_df[~spotify_df["Duration (ms)"].apply(lambda x: str(x).replace('.', '', 1).isdigit())]

# Display the problematic rows (if any)
invalid_popularity, invalid_duration


# Check for NaN values in numerical columns
spotify_df[["Popularity", "Duration (ms)"]].isna().sum()


# Generate pair plot without KDE to avoid potential issues
sns.pairplot(spotify_df[["Popularity", "Duration (ms)"]], diag_kind="hist", corner=True)
plt.show()


# Limit to top 10 artists with the most tracks for readability
top_10_artists = spotify_df["Artist"].value_counts().head(10).index
filtered_df = spotify_df[spotify_df["Artist"].isin(top_10_artists)]

# Box plot to explore track duration variation across artists
plt.figure(figsize=(12, 6))
sns.boxplot(x="Artist", y="Duration (ms)", data=filtered_df)
plt.xticks(rotation=45)
plt.title("Variation in Track Duration Across Top 10 Artists")
plt.show()


# Swarm plot to visualize track popularity distribution across top 10 artists
plt.figure(figsize=(12, 6))
sns.swarmplot(x="Artist", y="Popularity", data=filtered_df, size=4, alpha=0.7)
plt.xticks(rotation=45)
plt.title("Distribution of Track Popularity Across Top 10 Artists")
plt.show()
```
